```python
# Display-only pipeline: generate tables & charts WITHOUT saving any CSV/PNG files.
# - Reads /mnt/data/train.csv
# - Shows DataFrames inline (no caas_jupyter_tools)
# - Renders charts inline with matplotlib (no seaborn, one chart per figure)

import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# =========
# 0) Utils
# =========
def display_df(title: str, df: pd.DataFrame):
    """
    Try rich display in notebooks; otherwise print a nice text table.
    For terminal prettiness, 'pip install tabulate' (optional).
    """
    try:
        from IPython.display import display, HTML
        display(HTML(f"<h3>{title}</h3>"))
        display(df)
    except Exception:
        print(f"\n=== {title} ===")
        try:
            from tabulate import tabulate
            print(tabulate(df, headers="keys", tablefmt="github", showindex=False))
        except Exception:
            # Fallback to plain text
            print(df.to_string(index=False))



```

```python
# ==============================
# 1) Load & schema
# ==============================
df = pd.read_csv("train.csv")
assert "rule_violation" in df.columns, "Expected 'rule_violation' column"

# text column guess
TEXT_CANDIDATES = ["body", "text", "comment", "comment_text", "content"]
text_col = next((c for c in TEXT_CANDIDATES if c in df.columns), None)
assert text_col is not None, f"One of {TEXT_CANDIDATES} must exist in columns"

y = pd.to_numeric(df["rule_violation"], errors="coerce").fillna(0).astype(int)
t = df[text_col].fillna("").astype(str)

# Column summary
col_summary = []
for c in df.columns:
    s = df[c]
    col_summary.append({
        "column": c,
        "dtype": str(s.dtype),
        "n_missing": int(s.isna().sum()),
        "pct_missing": round(100*s.isna().mean(), 3),
        "n_unique": int(s.nunique()),
        "example": s.dropna().astype(str).head(1).tolist()[0] if s.notna().any() else ""
    })
col_summary = pd.DataFrame(col_summary).sort_values("pct_missing", ascending=False)
display_df("EDA â€” Column summary", col_summary)

# Target distribution
td = y.value_counts(dropna=False).rename_axis("rule_violation").reset_index(name="count")
td["percent"] = (td["count"]/len(df)*100).round(3)
display_df("EDA â€” Target distribution", td)

# Duplicates (overall & text)
dup_rows = int(df.duplicated().sum())
dup_text = int(t.duplicated().sum())
print(f"Duplicate rows: {dup_rows} | Duplicate texts: {dup_text}")
```

```python
# ==============================
# 2) Text features (sentiment-ish & meta)
# ==============================
def normalize(text: str) -> str:
    text = text.lower()
    text = re.sub(r"[^a-z0-9' ]+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

norm = t.apply(normalize)
tokens_list = norm.str.split()

# Lexicons
positive_words = {
    "good","great","awesome","amazing","love","like","nice","kind","helpful","thanks","thank","appreciate",
    "brilliant","fantastic","excellent","well","welldone","congrats","congratulations","cool","sweet","cheers",
    "solid","decent","smart","interesting","insightful","clear","accurate","agree","true","valid","correct"
}
negative_words = {
    "bad","worse","worst","hate","hates","hated","awful","terrible","horrible","disgusting","gross","stupid",
    "idiot","idiotic","moron","dumb","dumber","dumbest","trash","garbage","sucks","loser","pathetic","toxic",
    "annoying","angry","mad","ridiculous","bs","nonsense","spam","fake","liar","lying","lie","lies","ignorant",
    "ugly","useless","pointless","wrong","incorrect","false","misleading","cringe","cringy","wtf","wtheck",
    "shut","shutup","goaway","hell","nazi","racist","sexist","homophobic","offensive","rude","jerk"
}
profanity_words = {
    "fuck","fucking","fucked","shit","bullshit","bitch","bitches","ass","asses","asshole","dick","dicks","prick",
    "bastard","crap","cunt","piss","pissed","damn","goddamn","mf","motherfucker","retard","retarded","dumbass",
    "screw","screwyou"
}
neg_phrases = {"screw you","shut up","kill yourself","die in a hole","go to hell"}

def count_lex(tokens, lex):
    return sum(1 for w in tokens if w in lex)

def has_phrase(text, phrases):
    return int(any(p in text for p in phrases))

pos_cnt = tokens_list.apply(lambda toks: count_lex(toks, positive_words)).astype(int)
neg_cnt = tokens_list.apply(lambda toks: count_lex(toks, negative_words)).astype(int)
profanity_cnt = tokens_list.apply(lambda toks: count_lex(toks, profanity_words)).astype(int)
neg_phrase_flag = norm.apply(lambda s: has_phrase(s, neg_phrases))

char_len = t.str.len()
word_count = tokens_list.apply(len)
exclaim_cnt = t.str.count("!")
has_url = t.str.contains(r"http[s]?://", regex=True, case=False, na=False).astype(int)
has_mention = t.str.contains(r"u/|@[\w_]+", regex=True, na=False).astype(int)
upper_ratio = t.apply(lambda x: sum(1 for ch in x if ch.isupper())/max(1,len(x)))

sent_score = (pos_cnt - neg_cnt) / (1 + pos_cnt + neg_cnt)
sent_label = pd.cut(sent_score, bins=[-np.inf,-0.2,0.2,np.inf], labels=["negative","neutral","positive"])

text_stats = pd.DataFrame([{
    "rows_with_empty_text": int((t.str.strip()=="").sum()),
    "char_len_mean": float(char_len.mean()),
    "char_len_median": float(char_len.median()),
    "char_len_p95": float(char_len.quantile(0.95)),
    "word_count_mean": float(word_count.mean()),
    "word_count_median": float(word_count.median()),
    "url_rate_%": float(100*has_url.mean()),
    "mention_rate_%": float(100*has_mention.mean()),
    "upper_ratio_mean": float(upper_ratio.mean()),
    "exclaim_cnt_mean": float(exclaim_cnt.mean()),
}])
display_df("EDA â€” Text stats)", text_stats)

by_class = pd.DataFrame({
    "char_len": char_len,
    "word_count": word_count,
    "upper_ratio": upper_ratio,
    "exclaim_cnt": exclaim_cnt,
    "sent_score": sent_score,
    "rule_violation": y,
})
by_class_stats = by_class.groupby("rule_violation").agg({
    "char_len": ["mean","median", lambda s: s.quantile(0.9)],
    "word_count": ["mean","median", lambda s: s.quantile(0.9)],
    "upper_ratio": ["mean","median"],
    "exclaim_cnt": ["mean","median"],
    "sent_score": ["mean","median"],
})
by_class_stats.columns = ["_".join(filter(None, c)).replace("<lambda>","p90")
                          for c in by_class_stats.columns.values]
by_class_stats = by_class_stats.reset_index()
display_df("EDA â€” By-class text stats", by_class_stats)

```

```python
# ==============================
# 3) Sentiment vs Violation â€” tables & charts
# ==============================
# bucket table
tbl_counts = pd.crosstab(sent_label, y, rownames=["sentiment"], colnames=["rule_violation"])
vrate_by_bucket = (tbl_counts.div(tbl_counts.sum(axis=1), axis=0)
                   .rename(columns={0:"non_violation_rate",1:"violation_rate"})).reset_index()
display_df("Sentiment â€” bucket Ã— violation counts", tbl_counts.reset_index())
display_df("Sentiment â€” violation rate by bucket", vrate_by_bucket)

# decile curve
deciles = pd.qcut(sent_score, 10, labels=False, duplicates="drop")
vrate_by_decile = pd.DataFrame({"decile": deciles, "rule_violation": y}).groupby("decile")["rule_violation"].mean().reset_index()
display_df("Sentiment â€” violation rate by decile", vrate_by_decile)

# charts (one per figure; no color/style set)
plt.figure()
plt.bar(td["rule_violation"].astype(str), td["count"])
plt.title("Class Balance â€” rule_violation")
plt.xlabel("rule_violation"); plt.ylabel("Count")
plt.tight_layout(); plt.show()

plt.figure()
capped = np.minimum(char_len, np.percentile(char_len, 99))
plt.hist(capped, bins=40)
plt.title(f"Histogram of {text_col} character length (capped at 99th pct)")
plt.xlabel("char_len"); plt.ylabel("Frequency")
plt.tight_layout(); plt.show()

plt.figure()
plt.boxplot([sent_score[y==0], sent_score[y==1]], labels=["non_violation(0)", "violation(1)"])
plt.title("Sentiment score by rule_violation class")
plt.ylabel("sent_score")
plt.tight_layout(); plt.show()

plt.figure()
plt.plot(vrate_by_decile["decile"], vrate_by_decile["rule_violation"], marker="o")
plt.title("Violation rate vs. sentiment-score decile (lower=more negative)")
plt.xlabel("sentiment score decile"); plt.ylabel("violation_rate")
plt.tight_layout(); plt.show()

```

```python
# ==============================
# 4) Token-level signals (BoW + L1 logreg) â€” display only
# ==============================
vectorizer = CountVectorizer(lowercase=True, token_pattern=r"[A-Za-z']{2,}", min_df=5, binary=True)
X = vectorizer.fit_transform(t)
clf = LogisticRegression(penalty="l1", C=1.0, solver="liblinear", max_iter=300)
clf.fit(X, y)
auc = roc_auc_score(y, clf.decision_function(X))
print(f"AUC (BoW L1-LogReg, in-sample, display-only): {auc:.3f}")

vocab = np.array(vectorizer.get_feature_names_out())
coef = clf.coef_.ravel()
k = 25
top_pos_idx = np.argsort(coef)[-k:][::-1]
top_neg_idx = np.argsort(coef)[:k]

top_to_violation = pd.DataFrame({
    "token": vocab[top_pos_idx],
    "coef": coef[top_pos_idx],
    "odds_ratio": np.exp(coef[top_pos_idx]).round(3)
})
top_to_nonviolation = pd.DataFrame({
    "token": vocab[top_neg_idx],
    "coef": coef[top_neg_idx],
    "odds_ratio": np.exp(coef[top_neg_idx]).round(3)
})
display_df("Tokens â€” Toward violation", top_to_violation)
display_df("Tokens â€” Toward non-violation", top_to_nonviolation)

```

```python
# ==============================
# 5) Feature-only interpretable model â€” display only
# ==============================
feat = pd.DataFrame({
    "sent_score": sent_score,
    "pos_cnt": pos_cnt,
    "neg_cnt": neg_cnt,
    "profanity_cnt": profanity_cnt,
    "neg_phrase_flag": neg_phrase_flag,
    "char_len": char_len,
    "word_count": word_count,
    "exclaim_cnt": exclaim_cnt,
    "upper_ratio": upper_ratio,
    "has_url": has_url,
    "has_mention": has_mention,
})
clf_feat = LogisticRegression(max_iter=500)
clf_feat.fit(feat.values, y)
feat_coefs = pd.DataFrame({
    "feature": feat.columns,
    "coef": clf_feat.coef_.ravel(),
    "odds_ratio": np.exp(clf_feat.coef_.ravel()).round(3)
}).sort_values("coef", ascending=False).reset_index(drop=True)
display_df("Interpretable model â€” feature coefficients", feat_coefs)

```

```python
# ==============================
# 6) Contradiction samples â€” display only
# ==============================
sent_label_str = sent_label.astype(str)
pos_violate_mask = ((y==1) & (sent_label_str=="positive"))
neg_nonviolate_mask = ((y==0) & (sent_label_str=="negative"))

def trim(s, n=220):
    s = str(s).replace("\n"," ").strip()
    return s if len(s)<=n else s[:n-3]+"..."

def extract_samples(mask, topn=20, positive=True):
    ss = pd.DataFrame({
        "rule_violation": y[mask],
        "sent_label": sent_label_str[mask],
        "sent_score": sent_score[mask],
        "pos_cnt": pos_cnt[mask],
        "neg_cnt": neg_cnt[mask],
        "profanity_cnt": profanity_cnt[mask],
        "neg_phrase_flag": neg_phrase_flag[mask],
        "word_count": word_count[mask],
        "exclaim_cnt": exclaim_cnt[mask],
        "upper_ratio": upper_ratio[mask],
        "has_url": has_url[mask],
        "has_mention": has_mention[mask],
        "text": t[mask],
    }).copy()
    ss["text_preview"] = ss["text"].apply(trim)
    ss = ss.sort_values("sent_score", ascending=positive).head(topn) if not positive else \
         ss.sort_values("sent_score", ascending=False).head(topn)
    return ss

pos_violate_samples = extract_samples(pos_violate_mask, topn=20, positive=True)
neg_nonviolate_samples = extract_samples(neg_nonviolate_mask, topn=20, positive=False)
display_df("Samples â€” Positive sentiment but Violated", pos_violate_samples)
display_df("Samples â€” Negative sentiment but Non-violated", neg_nonviolate_samples)

```

```python
# ==============================
# 7) Heuristic rule-type labels & simple rules â€” display only
# ==============================
personal_attack = {"idiot","idiotic","moron","jerk","loser","stupid","dumb","dumbass","asshole","bastard","prick"}
hate_terms = {"nazi","racist","sexist","homophobic","retard","retarded"}
law_words = {"sue","lawyer","attorney","police","court","illegal","legal","legally","code","codes"}
mod_words = {"mod","mods","moderator","moderators","ban","banned","unban"}

personal_attack_cnt = tokens_list.apply(lambda toks: count_lex(toks, personal_attack)).astype(int)
hate_cnt = tokens_list.apply(lambda toks: count_lex(toks, hate_terms)).astype(int)
law_cnt = tokens_list.apply(lambda toks: count_lex(toks, law_words)).astype(int)
mod_cnt = tokens_list.apply(lambda toks: count_lex(toks, mod_words)).astype(int)

cats = {
    "profanity/harassment": (profanity_cnt>0) | (neg_phrase_flag==1) | (personal_attack_cnt>0),
    "hate/offensive": (hate_cnt>0),
    "legal-threat/dispute": (law_cnt>0),
    "moderation-targeting": (mod_cnt>0),
    "user-mention/targeting": (has_mention==1),
    "url/spam-like": (has_url==1),
    "shouting/emphasis": (upper_ratio>0.2) | (exclaim_cnt>=3),
}
multi = pd.DataFrame({k: v.astype(int) for k,v in cats.items()})
multi["neutral/other"] = (multi.sum(axis=1)==0).astype(int)

base_violation_rate = y.mean()
summary_rows = []
for k in multi.columns:
    mask = (multi[k]==1)
    n = int(mask.sum())
    v_rate = float(y[mask].mean()) if n>0 else np.nan
    summary_rows.append({
        "category": k,
        "n_rows": n,
        "share_%": round(100*n/len(df),2),
        "violation_rate": round(v_rate,4) if n>0 else None,
        "lift_vs_base": round((v_rate/base_violation_rate - 1.0),3) if (n>0 and base_violation_rate>0) else None
    })
cat_summary = pd.DataFrame(summary_rows).sort_values("violation_rate", ascending=False)
display_df("Rule types â€” category summary", cat_summary)

# simple rules evaluation (C1~C7)
cond_defs = {
    "C1_violation_profanity_or_threat": ( (profanity_cnt>0) | (neg_phrase_flag==1) | (personal_attack_cnt>0) ),
    "C2_nonviolation_positive_clean": ( (sent_score>0.3) & (profanity_cnt==0) & (neg_phrase_flag==0) & (exclaim_cnt<2) ),
    "C3_violation_legal_words": (law_cnt>0),
    "C4_violation_shouting": ( (exclaim_cnt>=3) | (upper_ratio>0.2) ),
    "C5_violation_url_short": ( (has_url==1) & (word_count<=12) ),
    "C6_violation_user_mention": (has_mention==1),
    "C7_nonviolation_neutral_clean": ( (sent_score.abs()<0.1) & (profanity_cnt==0) & (neg_phrase_flag==0) & (has_url==0) & (exclaim_cnt==0) ),
}

total = len(y)
tot_viol = int(y.sum())
base_rate = y.mean()

rows_eval = []
for name, mask in cond_defs.items():
    n = int(mask.sum())
    if n==0:
        continue
    viol = int(y[mask].sum())
    viol_rate = viol / n
    nonviol_rate = 1 - viol_rate
    if "violation" in name and "nonviolation" not in name:
        precision = viol_rate
        recall = viol / tot_viol if tot_viol>0 else np.nan
        fpr_like = (n - viol) / (total - tot_viol) if (total - tot_viol)>0 else np.nan
    else:
        precision = nonviol_rate
        recall = (n - viol) / (total - tot_viol) if (total - tot_viol)>0 else np.nan
        fpr_like = viol / tot_viol if tot_viol>0 else np.nan
    rows_eval.append({
        "condition": name,
        "coverage_n": n,
        "coverage_%": round(100*n/total,2),
        "violation_rate": round(viol_rate,4),
        "non_violation_rate": round(nonviol_rate,4),
        "lift_vs_base": round((viol_rate/base_rate - 1.0),3),
        "precision": round(precision,4),
        "recall": round(recall,4),
        "fpr_like": round(fpr_like,4),
    })

rules_eval = pd.DataFrame(rows_eval).sort_values("condition")
display_df("Simple rules â€” evaluation", rules_eval)

# chart: violation rate by rule-type
plt.figure()
plt.bar(cat_summary["category"], cat_summary["violation_rate"])
plt.title("Violation rate by heuristic rule-type")
plt.xticks(rotation=45, ha="right")
plt.ylabel("violation_rate")
plt.tight_layout(); plt.show()

# chart: violation rate by simple rules
plt.figure()
plt.bar(rules_eval["condition"], rules_eval["violation_rate"])
plt.title("Violation rate by simple rule condition")
plt.xticks(rotation=45, ha="right")
plt.ylabel("violation_rate")
plt.tight_layout(); plt.show()

print("Display-only run completed.")
```
